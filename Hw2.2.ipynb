```python
import pandas as pd
df = pd.read_csv('2017_jun_final.csv')
print(df.head())
df.shape
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python

df['Язык.программирования']
newdf = df.dropna(axis=1).copy()
newdf['Язык.программирования'] = df['Язык.программирования']
```

```python
newdf.isnull().sum()
```

```python
newdf.dropna()
newdf.shape
```

```python
python_data = newdf[newdf['Язык.программирования'] == 'Python'].copy()
python_data.shape
```

```python
grouped_by_position = python_data.groupby('Должность')
grouped_by_position.head()
```

```python
aggregated_data = python_data.groupby('Должность').agg({'Зарплата.в.месяц': ['min', 'max']})
aggregated_data
```

```python
def fill_avg_salary(row):
    avg_salary = row['Зарплата.в.месяц'].mean()
    return avg_salary


aggregated_data['avg'] = aggregated_data.apply(fill_avg_salary, axis=1)
aggregated_data
```

```python
describe_stats = aggregated_data['avg'].describe()
describe_stats
```

```python
aggregated_data.to_csv('aggregated_data_table.csv', index=False)
describe_stats.to_csv('describe_stats_table.csv', index=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
python_data['AverageSalary'] = python_data.groupby('Должность')['Зарплата.в.месяц'].transform('mean')

plt.figure(figsize=(12, 6))
sns.barplot(x='Должность', y='AverageSalary', data=python_data, errorbar=None)
plt.title('Average Salary by Profession')
plt.xlabel('Profession')
plt.ylabel('Average Salary')
plt.xticks(rotation=45)
plt.show()
```

```python
avg_salary_by_profession = python_data.groupby('Должность')['AverageSalary'].mean()


plt.figure(figsize=(6, 6))
plt.pie(avg_salary_by_profession, labels=avg_salary_by_profession.index, autopct='%1.1f%%', startangle=140)
plt.title('Average Salary Distribution by Profession')
plt.show()
```
