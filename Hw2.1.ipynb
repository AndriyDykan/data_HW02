```python
import pandas as pd
import numpy as np
http = "https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C_%D0%B2_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D1%96"
temp = pd.read_html(http,match = "Коефіцієнт народжуваності у регіонах України у 1950—2014 роках")

temp[0].set_index('регіон', inplace=True)
print(temp[0].head(20))
df = temp[0]
```

```python
df.shape
```

```python
df.replace("—",np.nan,inplace=True)
df
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
df = df.drop(df.index[-1])
df
```

```python
df.fillna(df.mean(),inplace = True)
df
```

```python
average = df.mean().mean()
print(average)
f_df = df[df['2014'] > average]['2014']
f_df
```

```python
df['2014'].idxmax()
print(f"{df['2014'].idxmax()} {df['2014'].max()}" )
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
df['2014'].sort_values().plot(kind='bar', color='skyblue')
plt.xlabel('Регіон')
plt.ylabel('Кількість народжень у 2014 році')
plt.title('Народжуваність по регіонах у 2014 році')
plt.show()
```

```python
import seaborn as sns
average_values = df.mean(axis=0, skipna=True)
sns.set(style="darkgrid",palette ='coolwarm')
# Plot 3: Line plot for average population values over the years
plt.figure(figsize=(12, 6))
sns.lineplot(x=average_values.index, y=average_values.values)
plt.title('Average Population Across Regions Over the Years')
plt.xlabel('Year')
plt.ylabel('Average Population')
plt.xticks(rotation=45)
plt.show()
```

```python
df.columns
```

```python
region_to_plot = 'Львівська область'

# Extract the relevant data for the scatter plot
years = df.columns[:-1]
population_data = df.loc[region_to_plot][:-1]

# Plot 4: Scatter plot for a specific region's population over the years
plt.figure(figsize=(12, 6))
sns.scatterplot(x=years, y=population_data, marker='o', color='purple', s=100)
plt.title(f'{region_to_plot} Population Over the Years')
plt.xlabel('Year')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.stripplot(x="1990", y=df.index, data=df, jitter=True, marker='o', color='green', size=8)
plt.title('Population Distribution Across Regions in 1990')
plt.xlabel('Population')
plt.ylabel('Region')
plt.show()
```
